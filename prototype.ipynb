```python
arr_neurons = ['SPK05a.1', 'SPK12a.1','SPK13a.1','SPK14a.1']
datadict = {
    'SPK05a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK12a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK13a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK14a.1': {
        'std': [],
        'mean': [],
        'w': []
    }
}
for j in arr_neurons:
    # finding good width for each neuron
    sample_spikes = df[j].dropna()
    comparison_matrix = []
    sample_behavior_arr = []
    arr1 = []
    std_arr = [0]*20
    for w in np.linspace(.5,5,20):
        sample_behavior_arr = [0]*len(interactions)
        for interaction_cnt in range(0,len(interactions)):
            
            s2_arr = []
            sample_behavior_1 = sample_spikes[abs(sample_spikes-interactions[interaction_cnt]) < w]
            sample_behavior_2 = sample_behavior_1
            s2_arr.append(sample_behavior_2)
            sample_behavior_1.index = pd.to_datetime(sample_behavior_1, unit='s')
            first_data_sec = (sample_behavior_1.index[0].minute*60 + sample_behavior_1.index[0].second + sample_behavior_1.index[0].microsecond/1000000)
            last_data_sec = (sample_behavior_1.index[-1].minute*60 + sample_behavior_1.index[-1].second + sample_behavior_1.index[-1].microsecond/1000000)

            num_zeros_before = (math.floor((first_data_sec-interactions[interaction_cnt]+w)/.05))
            num_zeros_after = (math.floor((interactions[interaction_cnt]+w-last_data_sec)/.05))

            sample_behavior_1 = sample_behavior_1.groupby(pd.Grouper(freq='50ms')).count()

            sample_behavior_1 = np.concatenate(([0]*num_zeros_before, sample_behavior_1.values,[0]*num_zeros_after), axis=None)
            sample_behavior_arr[interaction_cnt] = sample_behavior_1
            
        sample_behavior_arr = np.transpose(sample_behavior_arr)

                
        std_arr1 = []
        for x in range(len(interactions)):
            y = np.std(sample_behavior_arr[x])
            std_arr1.append(y)

        
        mean_arr1 = []
        for x in range(len(interactions)):
            y = np.mean(sample_behavior_arr[x])
            mean_arr1.append(y)
            
        print("mean_arr ", mean_arr1)
        


        print('std= ', std_arr)
        plt.title(j)
        pd.DataFrame(np.linspace(.5,5,20),std_arr1).plot()
        
        datadict[j]["std"] = std_arr1
        datadict[j]["mean"] = mean_arr1
        datadict[j]['w'] = w

print('data =',datadict)
```

```python
errors = []

for width in np.linspace(0, 20, 100):
    n_bars = 10
    period = '%dms' % (math.floor(abs(2000*width/n_bars))+1)
    #print('period:', period)
    sample_behavior_1 = sample_spikes[abs(sample_spikes-interactions[0]) < width]
    sample_behavior_2 = sample_spikes[abs(sample_spikes-interactions[1]) < width]
    sample_behavior_1.index = pd.to_datetime(sample_behavior_1, unit='s')
    sample_behavior_2.index = pd.to_datetime(sample_behavior_2, unit='s')
    sample_behavior_1_grouped = sample_behavior_1.groupby(pd.Grouper(freq=period)).count()
    sample_behavior_2_grouped = sample_behavior_2.groupby(pd.Grouper(freq=period)).count()

    a = sample_behavior_1_grouped.reset_index(drop=True)
    b = sample_behavior_2_grouped.reset_index(drop=True)
    error = abs(a-b).mean()
    print(a.shape, b.shape)
    print('width:', width, '; error:', error)
    errors.append(error/width)
    
pd.DataFrame(errors).plot()
```
