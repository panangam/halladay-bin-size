```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np

df = pd.read_excel('sampledata_elissa.xlsx', sheet_name='Sheet1', header=2)
df
```

```python
interactions = df['Avert'].dropna()
interactions
```

```python
sample_spikes = df['SPK05a.1'].dropna()
sample_spikes
```

```python
width=20
sample_behavior = sample_spikes[abs(sample_spikes-interactions[1]) < width]
sample_behavior.index = pd.to_datetime(sample_behavior, unit='s')
pd.DataFrame(sample_behavior.groupby(pd.Grouper(freq='500ms')).count()).plot(kind='bar')
```

```python
arr_neurons = ['SPK05a.1', 'SPK12a.1','SPK13a.1','SPK14a.1']
datadict = {
    'SPK05a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK12a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK13a.1': {
        'std': [],
        'mean': [],
        'w': []
    },
    'SPK14a.1': {
        'std': [],
        'mean': [],
        'w': []
    }
}
for j in arr_neurons:
    sample_spikes = df[j].dropna()
    comparison_matrix = []
    std_arr = []
    for w in np.linspace(.5,10,20):
        for interaction_cnt in range(0,len(interactions)):
            sample_behavior_1 = sample_spikes[abs(sample_spikes-interactions[interaction_cnt]) < w]
            sample_behavior_1.index = pd.to_datetime(sample_behavior_1, unit='s')
            sample_behavior_1 = sample_behavior_1.groupby(pd.Grouper(freq='100ms')).count()
#           std_arr.append(sample_behavior_1.std(axis= None, skipna=True, level=None, ddof=1, numeric_only=None))
#           print('sample behavior =', sample_behavior_1)
            comparison_matrix.append(sample_behavior_1)
                
        std_arr1 = []
        for x in range(len(comparison_matrix)):
            y = comparison_matrix[x].std(axis= None, skipna=True, level=None, ddof=1, numeric_only=None)
            std_arr1.append(y)        
        std_arr = pd.DataFrame(index=np.linspace(.5,10,20),data=std_arr1)
        
        mean_arr1 = []
        for x in range(len(comparison_matrix)):
            y = comparison_matrix[x].mean()
            mean_arr1.append(y)
            
        print("mean_arr ", mean_arr1)
        


        print('std= ', std_arr)
        print('mat=', comparison_matrix)

        pd.DataFrame(np.linspace(.5,10,20),std_arr).plot()
        
        datadict[j]["std"] = std_arr
        datadict[j]["mean"] = mean_arr1

print(datadict)
```

```python
errors = []

for width in np.linspace(0, 20, 100):
    n_bars = 10
    period = '%dms' % (math.floor(abs(2000*width/n_bars))+1)
    #print('period:', period)
    sample_behavior_1 = sample_spikes[abs(sample_spikes-interactions[0]) < width]
    sample_behavior_2 = sample_spikes[abs(sample_spikes-interactions[1]) < width]
    sample_behavior_1.index = pd.to_datetime(sample_behavior_1, unit='s')
    sample_behavior_2.index = pd.to_datetime(sample_behavior_2, unit='s')
    sample_behavior_1_grouped = sample_behavior_1.groupby(pd.Grouper(freq=period)).count()
    sample_behavior_2_grouped = sample_behavior_2.groupby(pd.Grouper(freq=period)).count()

    a = sample_behavior_1_grouped.reset_index(drop=True)
    b = sample_behavior_2_grouped.reset_index(drop=True)
    error = abs(a-b).mean()
    print(a.shape, b.shape)
    print('width:', width, '; error:', error)
    errors.append(error/width)
    
pd.DataFrame(errors).plot()
```
